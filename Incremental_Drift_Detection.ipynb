```python
from skmultiflow.drift_detection.adwin import ADWIN
import pandas as pd
from skmultiflow.data.random_rbf_generator_drift import RandomRBFGeneratorDrift
```

```python
# Incremental drift is generated using Random RBF generator
stream = RandomRBFGeneratorDrift(model_random_state=99, sample_random_state = 50,
 n_classes = 2, n_features = 5, n_centroids = 50, change_speed=0.01,
 num_drift_centroids=50)
```

```python
X,Y= stream.next_sample(500)
```

```python
X[0]
```

```python
list(X[0])
```

```python
adwin=ADWIN()
for i in range(len(X)):
  adwin.add_element(X[i][0])
  adwin.add_element(X[i][1])
  adwin.add_element(X[i][2])
  adwin.add_element(X[i][3])
  adwin.add_element(X[i][4])
  if adwin.detected_change():
    print("Change Detected in data at index- "+str(i))
```

```python
from skmultiflow.drift_detection.eddm import EDDM
eddm = EDDM()
for i in range(len(X)):
  eddm.add_element(X[i][0])
  eddm.add_element(X[i][1])
  eddm.add_element(X[i][2])
  eddm.add_element(X[i][3])
  eddm.add_element(X[i][4])
  if(eddm.detected_warning_zone()):
    print("Warning Detected in data at index- "+str(i))
  if(eddm.detected_change()):
    print("Change Detected in data at index- "+str(i))
```

```python
from skmultiflow.drift_detection.hddm_w import HDDM_W
hddm_w = HDDM_W()
for i in range(len(X)):
  hddm_w.add_element(X[i][0])
  hddm_w.add_element(X[i][1])
  hddm_w.add_element(X[i][2])
  hddm_w.add_element(X[i][3])
  hddm_w.add_element(X[i][4])
  if(hddm_w.detected_warning_zone()):
    print("Warning Detected in data at index- "+str(i))
  if(hddm_w.detected_change()):
    print("Change Detected in data at index- "+str(i))
```
