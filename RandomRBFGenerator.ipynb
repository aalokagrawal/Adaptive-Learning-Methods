```python
from skmultiflow.data.random_rbf_generator_drift import RandomRBFGeneratorDrift
```

```python
stream = RandomRBFGeneratorDrift(model_random_state=99, sample_random_state = 50,
 n_classes = 2, n_features = 5, n_centroids = 50, change_speed=0.01,
 num_drift_centroids=50)
```

```python
import pandas as pd
X,Y= stream.next_sample(500)
```

```python
stream.n_remaining_samples()
```

```python
df= pd.DataFrame(X)
df["labels"]=Y
```

```python
df.to_csv("/content/drive/MyDrive/Geneator_data/RBFGeneratorData",index=False)
```
