```python
data = pd.read_csv("/content/drive/MyDrive/Real_world_data/NEweather_data.csv",index_col=None,header=None)
```

```python
from skmultiflow.drift_detection.adwin import ADWIN
adwin=ADWIN()
cnt=0
for i in range(len(data)):
  X= data.iloc[i]
  X=list(X)
  for j in range(len(X)):
    adwin.add_element(X[j])
  if adwin.detected_change():
    print("Change Detected in data at index- "+str(i))
    cnt+=1
print("Total no. of drift detected "+str(cnt))
```

```python
from skmultiflow.core.base import BaseEstimator
from skmultiflow.drift_detection.eddm import EDDM
eddm = EDDM()
cnt=0
for i in range(len(data)):
  X=data.iloc[i]
  X=list(X)
  for j in range(len(X)):
    eddm.add_element(X[j])
  if(eddm.detected_warning_zone()):
    print("Warning Detected in data at index- "+str(i))
  if(eddm.detected_change()):
    print("Change Detected in data at index- "+str(i))
    cnt+=1
print("Total no. of drift detected "+str(cnt))
```

```python
from skmultiflow.drift_detection.hddm_w import HDDM_W
hddm_w = HDDM_W()
cnt=0
for i in range(len(data)):
  X=data.iloc[i]
  X=list(X)
  for j in range(len(X)):
    hddm_w.add_element(X[j])
  if(hddm_w.detected_warning_zone()):
    print("Warning Detected in data at index- "+str(i))
  if(hddm_w.detected_change()):
    print("Change Detected in data at index- "+str(i))
    cnt+=1
print("Total no. of drift detected "+str(cnt))
```
